```python
import pandas as pd
import geopandas as gpd
import numpy as np
from datetime import datetime
import plotly.express as px

from urllib.request import urlopen
import json
```

```python
# handlig dates
dateparse = lambda x: datetime.strptime(x, '%d/%m/%Y')
```

```python
# Import CSV covid cases from local 
cc_csv = f"C:/capes_sql/post_covid/casos_cov_conf.csv"
cc_df = pd.read_csv(cc_csv, sep=',',parse_dates=['TipusCasData'], date_parser=dateparse)
cc_df.head(5)
```

```python
# NaN values in col MunicipiCodi as 0
cc_df['MunicipiCodi']= cc_df['MunicipiCodi'].fillna(0)
# col MunicipiCodi values as int
#cc_df['MunicipiCodi'] = cc_df['MunicipiCodi'].astype(int)
cc_df.head(3)
```

```python
cc_df = cc_df.groupby(["MunicipiCodi"])["NumCasos"].sum()
```

```python
cc_df = cc_df.to_frame() # serie to df
cc_df = cc_df.reset_index() # reset index
```
